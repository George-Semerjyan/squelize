```python
import sqlalchemy
from sqlalchemy import create_engine, MetaData
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Numeric, Text, Float
import pandas as pd
```

```python
measurments_csv = pd.read_csv("C:\\Users\Gevork\Desktop\clean_hawaii_measurements.csv")
measurments_csv
```

```python
stations_csv = pd.read_csv("C:\\Users\Gevork\Desktop\clean_hawaii_stations.csv")
stations_csv
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")
conn = engine.connect()
Base = declarative_base()
```

```python
class Measurment (Base):
    __tablename__ = "measurments"
    id = Column(Integer, primary_key=True)
    station = Column(String(55))
    date = Column(String(10))
    prcp = Column(Float)
    tobs = Column(Integer)
```

```python
class Station (Base):
    __tablename__ = "stations"
    id = Column(Integer, primary_key=True)
    station = Column(String(15))
    name = Column(String(100))
    latitude = Column(Float)
    longitude = Column(Float)
    elevation = Column(Float)
```

```python
Base.metadata.create_all(engine)
```

```python
measurment_data = measurments_csv.to_dict(orient='records')
station_data = stations_csv.to_dict(orient='records')
```

```python
print(measurment_data)
```

```python
print(station_data)
```

```python
metadata = MetaData(bind=engine)
metadata.reflect()
```

```python
stations_table = sqlalchemy.Table('stations', metadata, autoload=True)
measurments_table = sqlalchemy.Table('measurments', metadata, autoload=True)
```

```python
conn.execute(measurements_table.delete())
```

```python
conn.execute(stations_table.delete())
```

```python
conn.execute(stations_table.insert(), station_data)
conn.execute(measurments_table.insert(), measurment_data)
```
